```python
# pivot() method is used to convert the columns as column header
import pandas as pd
data=pd.read_csv('salesmen.csv')
```

```python
data.head(3)
```

```python
data.set_index('Date')
```

```python
data['Salesman'].nunique()
```

```python
data.nunique()
```

```python
data.pivot(index='Date',columns='Salesman',values='Revenue').head(5)
```

```python
# pivot_table() method is used to perform ms excel type operations in data frame
data.pivot_table(index='Date', values="Revenue",aggfunc='mean').head(2)
```

```python
data=pd.read_csv('foods.csv')
```

```python
data.head(2)
```

```python
# index parameter tell which column to used as grouping purpose
data.pivot_table(values='Spend',index='Gender',aggfunc='sum')
```

```python
data.pivot_table(values='Spend',index=['Gender','City','Item'],aggfunc="max")
```

```python
# columns parameter denotes which column set as column of new data
data.pivot_table(values='Spend',index=['City','Item'],columns='Gender',aggfunc='sum').head(3)
```

```python
# pd.melt() mehod perform is reverse type of pivot_table
data=pd.read_csv('quarters.csv')
```

```python
pd.melt(data,id_vars='Salesman',var_name="quater",value_name='Revenue')
```
