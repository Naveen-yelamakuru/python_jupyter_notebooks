```python
import pandas as pd
data=pd.read_csv('fortune1000.csv',index_col='Rank')
data.head(3)
```

```python
# group by using sector column
df=data.groupby('Sector')
```

```python
len(data)
```

```python
len(df)
```

```python
# it disply unique rows of df and also display the size of each group
df.size()
```

```python
data['Sector'].value_counts()
```

```python
# it diplays first single row of each group
df.first()
# it displays last single row of each group
df.last()
```

```python
# it disply all rows in index wise of each group 
df.groups
```

```python
# Retrieve a group with the .get_group()method.it display each group based on row value of grouping column
df.get_group('Retailing')
df.get_group('Energy').head(3)
```

```python
# methods on the groupby object and dataframe columns
df.max().head(1)
# in max method find maximum value on required column defaulty max parameter is left column of dataframe

```

```python
# it apply only numaric value rows,it display sum of all rows in column wise
df.sum().head(1)
df.mean()
df.min().head(1)
```

```python
# to perform the above operation on selected columns
df['Profits'].sum().head()
df['Profits','Revenue'].sum().head()
```

```python
# agg() method help to operform different operation on different columns in data frame in single line of syntax
df.agg({'Revenue':'sum','Profits':'mean'}).head(3)
```

```python
df.agg(['size','sum','mean']).head(3) # it helps to set sum column of numaric columns of a data frame
```

```python
# it helps to single row contain multple operations
df.agg({"Profits":['sum','mean','max']}).head(3)
```

```python
# Create a new data frame from original data frame with required rows
subdf=pd.DataFrame(columns=data.columns)
subdf
```

```python
# add required fields to the new dataframe
for sector,data in df:
    values=data.nlargest(1,'Profits')
    subdf.append(values)
```
