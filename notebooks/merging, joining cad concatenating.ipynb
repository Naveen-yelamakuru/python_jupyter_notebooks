```python
pd.concat(objs: it define names of data frames
    axis=0,0 defines rows and 1 defines columns
    join='outer', it what type of joining
    ignore_index: bool = False,at time combining two data frames both indexs of data frames also display to ingore that using this parameter
    keys=None,it display index in multi index type or it define columns names which is used for concat operation
    levels=None,
    names=None,
    verify_integrity: bool = False,
    sort: bool = False,
    copy: bool = True,)
```

```python
pd.concat(objs=[week1,week2])
```

```python
pd.concat(objs=[week1,week2],ignore_index=True,sort=True)
```

```python
# to performe join opertations in dataframe using merge() method
# performe to shared common values
week1.merge(Signature:
week1.merge(
    right, it defines another dataframe we want to merge
    how='inner',it define what type of join operation is performed
    on=None,it define which column is used for join operation
    left_on=None,it define first data frame column use for join operation
    right_on=None,it define second data frame colmn use for join operation
    left_index=False, it defing perform join operation based on index values
    right_index=False,
    sort=False,
    suffixes=('_x', '_y'), change names of new data frames
    copy=True,
    indicator=False,it help to indicate nan values coming from in which data frame
    validate=None,)
```

```python
week1.merge(week2,how='inner',on='Customer ID',suffixes=('_week1', '-week2'),)
```

```python
week1.merge(week2,how='inner',on='Food ID',suffixes=('_week1', '-week2'),)
```

```python
# perform inner join operation in multiple columns 
week1.merge(week2,how='inner',on=['Customer ID','Food ID'],suffixes=('_week1', '-week2'),)
# this syntax is equal to the week1.merge(week2) syntax
```

```python
# perform the outer join operation
merged=week1.merge(week2,how='outer',on='Customer ID',suffixes=('_week1', '-week2'),)
```

```python
# to indicate nan values came form which data frame
merged=week1.merge(week2,how='outer',on='Customer ID',suffixes=('_week1', '-week2'),indicator=True)
```
