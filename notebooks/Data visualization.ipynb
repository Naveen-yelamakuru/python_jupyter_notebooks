```python
import pandas as pd
from pandas_datareader import data
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
source=data.DataReader(name='BB',
    data_source='yahoo',
    start='2007-01-01',
    end='2020-12-31').head(3)
```

```python
source.plot()
```

```python
# create bar graphs
def operation(stack_price):
    if stack_price <=10:
        return "Poor"
    if stack_price <=50:
        return "good"
    else:
        return 'best'
```

```python
source['Close'].apply(operation).value_counts().plot(kind='bar')
```
